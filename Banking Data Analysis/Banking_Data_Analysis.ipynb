```python
#Examine the distribution of unique cataegories in categorical colunms
categorical_cols= df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for'{col}':")
  display(df[col].value_counts())
```

##Univariate Analysis

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor)
```

##Bivariate Analysis

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
#Histplot of value counts for different Occupation

for col in categorical_cols:
  if col=="Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```

##Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loan','Bank Deposit','Checking Accounts','Saving Accounts','Foreign Currency Account ']

# Clean column names
df.columns = df.columns.str.strip()

# Keep only valid column names that exist in the DataFrame
numerical_cols = [col for col in numerical_cols if col in df.columns]

# Plot
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Counts')
    plt.tight_layout()
    plt.show()
```

##Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loan', 'Bank Deposit', 'Checking Accounts','Saving Accounts', 'Foreign Currency Account']

# Clean column names
df.columns = df.columns.str.strip()

# Filter only those columns that actually exist
numerical_cols = [col for col in numerical_cols if col in df.columns]


correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```
